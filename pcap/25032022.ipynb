```python
class Alpha:
    value = "Alpha"
    def say(self):
        return self.value.lower()

class Beta(Alpha):
    value = "Beta"

class Gamma(Alpha):
    def say(self):
        return self.value.upper()

class Delta(Gamma, Beta):
    pass
d = Delta()
b = Beta()
```
