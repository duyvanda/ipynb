```python
df_raw1 = df_raw.iloc[:, 0:(ix_nmv+num_days)].copy()
df_raw1=df_raw1.drop(['NMV Daily', 'Total'], axis = 1)
df_nmv = df_raw1.melt(id_vars=['Group Brand','brand','platform'], var_name='date', value_name='nmv_target')
df_nmv
```

```python
df_raw2 = df_raw.iloc[:, ix_order:(ix_order+num_days)].copy()
df_raw2['ABC'] = 'Duy'
df_order = df_raw2.melt(id_vars=['ABC'], var_name='date', value_name='order_target')
df_order.drop(['ABC','date'],axis=1,inplace = True)
df_order
```

```python
df_raw3 = df_raw.iloc[:, ix_calendar:(ix_calendar+num_days)].copy()
df_raw3['ABC'] = 'Duy'
df_daytype = df_raw3.melt(id_vars=['ABC'], var_name='date', value_name='day_type')
df_daytype.drop(['ABC','date'],axis=1,inplace = True)
df_daytype
```

```python
df_raw4 = df_raw.iloc[:, ix_traffic:(ix_traffic+num_days)].copy()
df_raw4['ABC'] = 'Duy'
df_traffic = df_raw4.melt(id_vars=['ABC'], var_name='date', value_name='traffic_target')
df_traffic.drop(['ABC','date'],axis=1,inplace = True)
df_traffic.fillna(0,inplace = True)
df_traffic
```

```python
df_raw5 = df_raw.iloc[:, ix_fulfillment:(ix_fulfillment+num_days)].copy()
df_raw5['ABC'] = 'Duy'
df_fulfillment = df_raw5.melt(id_vars=['ABC'], var_name='date', value_name='fulfillment_model')
df_fulfillment.drop(['ABC','date'],axis=1,inplace = True)
df_fulfillment
```

```python
df = pd.concat([df_nmv,df_order,df_traffic,df_daytype,df_fulfillment],axis=1)
df['brand_platform'] = df['brand'] + df['platform']
df.drop(['Group Brand'],axis=1,inplace = True)
df = df.apply(lambda x: x.str.strip()).replace('', np.nan)
df
```

```python
df['nmv_target'] = df['nmv_target'].fillna(0).str.replace(',','')
df['order_target'] = df['order_target'].fillna(0).str.replace(',','')
df['traffic_target'] = df['traffic_target'].fillna(0).str.replace(',','')
```

```python
df['day_type'] = df['day_type'].fillna('BL')
df['nmv_target'] = pd.to_numeric(df['nmv_target'],errors='coerce').fillna(0).astype('int64')
df['order_target'] = pd.to_numeric(df['order_target'],errors='coerce').fillna(0).astype('int64')
df['traffic_target'] = pd.to_numeric(df['traffic_target'],errors='coerce').fillna(0).astype('int64')
df['fulfillment_model'] = df['fulfillment_model'].fillna('None')
df['date'] = pd.to_datetime(df['date'])
df = df.dropna(subset=["brand", "platform", "date", "day_type"]).copy()
```

```python
def get_postgres_engine(postgres_conf):
    engine = create_engine(
        "postgresql://{user}:{pw}@{host}:{port}/{db}".format(
            user=postgres_conf["user"],
            pw=postgres_conf["password"],
            host=postgres_conf["host"],
            port=postgres_conf["port"],
            db=postgres_conf["database"],
        )
    )
    return engine

postgres_conf = {
  "host": "opwh.cybjleasxg2d.ap-southeast-1.rds.amazonaws.com",
  "port": 5432,
  "user": "dieuanh",
  "password": "m!f3QKXdZQ1sXMPk",
  "database": "warehouse",
}

def insert_df_to_postgres(postgres_conf, tbl_name, df, primary_keys: list, schema="public"):
    if df.shape[0] < 1:
        return

    def norm_value(d: dict):
        for k, v in d.items():
            if isinstance(v, (list, tuple)):
                continue
            if pd.isnull(v):
                d[k] = None
        return d

    engine = get_postgres_engine(postgres_conf)
    meta = MetaData(schema=schema)

    conn = engine.connect()
    my_table = Table(tbl_name, meta, autoload=True, autoload_with=conn)
    insert_statement = postgresql.insert(my_table).values(
        list(map(norm_value, df.to_dict(orient="records")))
    )
    if len(primary_keys):
        upsert_statement = insert_statement.on_conflict_do_update(
            index_elements=primary_keys,
            set_={
                c.key: c for c in insert_statement.excluded if c.key not in primary_keys
            },
        )
        conn.execute(upsert_statement)
    else:
        conn.execute(insert_statement)
```
